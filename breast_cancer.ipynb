```python
# as we can see from above violinplots that there are some features which are closely related to each other 
#therefore we will use jointplots to check the similarity
import scipy.stats as stats
plt.figure(figsize =(10,10))
sn.jointplot(df.loc[:,'concavity_worst'], df.loc[:,'concave points_worst'], kind="regg", color="#ce1414").annotate(stats.pearsonr)
```

```python
#to compare 3 or more features together we use pair grid plot
df = df.loc[:,['radius_worst','perimeter_worst','area_worst']]
g = sn.PairGrid(df, diag_sharey=False)

g.map_lower(sn.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter)
g.map_diag(sn.kdeplot, lw=3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix , f1_score
from sklearn.metrics import accuracy_score 
```

```python
model = RandomForestClassifier(random_state = 43 )
```

```python
x_train , x_test , y_train , y_test = train_test_split(w , y , test_size = 0.3 , random_state = 42)
model = model.fit(x_train, y_train)
ac = accuracy_score(y_test , model.predict(x_test))
print("accuracy is : " + str(ac))
cm = confusion_matrix(y_test,model.predict(x_test))
sn.heatmap(cm,annot=True,fmt="d")
```
